```python
import os
import shutil
import pandas as pd
```

```python
audio_path = 'D:/ML/ESC-50/ESC-50-master/audio/'
metadata_path = 'D:/ML/ESC-50/ESC-50-master/meta/esc50.csv'
```

```python
df = pd.read_csv(metadata_path)
```

```python
focus_classes = ['keyboard_typing', 'mouse_click']
distracted_classes = ['siren', 'car_horn', 'laughing', 'clapping']
```

```python
os.makedirs('data/focus', exist_ok=True)
os.makedirs('data/distracted', exist_ok=True)
```

```python
for index, row in df.iterrows():
    if row['category'] in focus_classes:
        shutil.copy(audio_path + row['filename'], 'data/focus/')
    elif row['category'] in distracted_classes:
        shutil.copy(audio_path + row['filename'], 'data/distracted/')    
        
print("Data is sorted.")
```
